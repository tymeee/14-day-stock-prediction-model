```python
cont_vars = ['Open', 'High', 'Low', 'Close', 'Adj Close','Volume', 'rsi','stochastic','signal-stochastic','cci', 'williams%R','roc','awesome-oscillator','ultimate-oscillator','trix','KAMA','stochrsi','sma','macd','ema','adx','psar-up','psar-down', 'ichimoku_a','ichimoku_b','supertrend','KST','plus_di','minus_di','VI-dff','VI+','VI-','atr','bb_bbm',
              'donchianwidth','keltnerwidth','ulcer','chaikin_volatility','historical_volatility','upper_band','lower_band'] # Remove 'Ticker' from cont_vars
seq_length = 20
forecast_horizon = 14
target_col = 'return_14d'
ticker_col = 'Ticker'
date_col = 'Date'
batch_size = 64  # You can adjust this
```

```python
import numpy as np
import pandas as pd

def prepare_sequences_targets(df_input, seq_length, forecast_horizon, cont_vars, target_col, ticker_col, date_col, valid_pct=0.2):
    """
    Prepares time series sequences and targets from a DataFrame,
    splitting data into train/validation based on a time-based cutoff.
    Valid_pct = 0.0 means use all data as 'train' (or in this case, test).
    """
    # Ensure datetime
    if not np.issubdtype(df_input[date_col].dtype, np.datetime64):
        df_input[date_col] = pd.to_datetime(df_input[date_col])

    # Sort by ticker and date
    df_input = df_input.sort_values([ticker_col, date_col]).reset_index(drop=True)

    rows = []
    targets = []
    dates = []
    # Use a list comprehension to filter tickers that might not have enough data
    tickers_with_enough_data = [
        ticker for ticker in df_input[ticker_col].unique()
        if len(df_input[df_input[ticker_col] == ticker]) >= seq_length + forecast_horizon
    ]

    if not tickers_with_enough_data:
        print(f"Warning: No tickers have enough data ({seq_length + forecast_horizon} rows minimum) to form sequences.")
        # Return empty arrays immediately
        return (np.array([]), np.array([])), (np.array([]), np.array([]))


    for ticker in tickers_with_enough_data:
        group = df_input[df_input[ticker_col] == ticker].reset_index(drop=True)

        # Ensure required columns exist and handle potential NaNs/Infs in source data
        required_cols = cont_vars + [target_col]
        if not all(col in group.columns for col in required_cols):
             print(f"Warning: Missing required columns for ticker {ticker}. Skipping.")
             continue

        # Replace infinities before converting to numpy - assuming NaNs are handled later
        group = group.replace([np.inf, -np.inf], np.nan)

        data_arr = group[cont_vars].to_numpy(dtype=np.float32)
        target_arr = group[target_col].to_numpy(dtype=np.float32)
        date_arr = group[date_col].to_numpy()

        n = len(group)
        max_start = n - seq_length - forecast_horizon + 1

        if max_start <= 0:
            print(f"Warning: Ticker {ticker} has {n} rows, which is not enough for seq_length {seq_length} and forecast_horizon {forecast_horizon}.")
            continue # Skip this ticker

        for start in range(max_start):
            seq_end = start + seq_length
            target_idx = seq_end + forecast_horizon - 1

            # Basic bounds check (should be covered by max_start but good practice)
            if target_idx >= n:
                 # This shouldn't happen with correct max_start logic
                 print(f"Error: Index out of bounds for ticker {ticker} at start {start}, target_idx {target_idx}.")
                 continue

            # Check for NaN target
            if np.isnan(target_arr[target_idx]):
                continue

            seq_features = data_arr[start:seq_end].flatten()

            # Check for NaN in sequence features
            if np.isnan(seq_features).any():
                continue

            # If all checks pass, append the sequence, target, and date
            rows.append(seq_features)
            targets.append(target_arr[target_idx])
            dates.append(date_arr[target_idx])

    X = np.array(rows)
    y = np.array(targets, dtype=np.float32)
    dates = np.array(dates)

    if len(X) == 0:
        print("Warning: No valid sequences generated after filtering.")
        return (np.array([]), np.array([])), (np.array([]), np.array([]))


    # Split by time using dates array
    if valid_pct > 0 and len(dates) > 0:
        # Only calculate quantile if valid_pct > 0 and dates array is not empty
        try:
            # Convert to int64 for quantile calculation, then back to datetime
            dates_int = dates.astype('datetime64[s]').astype(np.int64)
            # Ensure q is between 0 and 1 for numpy.quantile
            q_value = 1 - valid_pct
            # Clamp q_value between 0 and 1 to avoid errors
            q_value = max(0.0, min(1.0, q_value))

            cutoff_date_int = np.quantile(dates_int, q_value)
            cutoff_date = np.datetime64(int(cutoff_date_int), 's')

            train_mask = dates < cutoff_date
            valid_mask = dates >= cutoff_date

            X_train, y_train = X[train_mask], y[train_mask]
            X_valid, y_valid = X[valid_mask], y[valid_mask]

            print(f"Train samples: {len(X_train)}, Valid samples: {len(X_valid)}")

            return (X_train, y_train), (X_valid, y_valid)

        except Exception as e:
            print(f"Error during quantile calculation for time split: {e}")
            # Return empty split if quantile calculation fails
            return (np.array([]), np.array([])), (np.array([]), np.array([]))

    else: # valid_pct is 0 or dates array is empty
        # No split, return all data as train/test
        print(f"Valid_pct is 0 or no dates available. Using all {len(X)} samples for training/testing.")
        return (X, y), (np.array([]), np.array([])) # Return empty validation set

```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib
import os
seq_length = 10
forecast_horizon = 1
cont_vars = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume','rsi','stochastic','signal-stochastic','cci', 'williams%R','roc','awesome-oscillator','ultimate-oscillator','trix','KAMA','stochrsi','sma','macd','ema','adx','psar-up','psar-down', 'ichimoku_a','ichimoku_b','supertrend','KST','plus_di','minus_di','VI-dff','VI+','VI-','atr','bb_bbm',
              'donchianwidth','keltnerwidth','ulcer','chaikin_volatility','historical_volatility','upper_band','lower_band'] # Remove 'Ticker' from cont_vars
target_col = 'return_14d'
ticker_col = 'Ticker'
date_col = 'Date'
valid_pct = 0.2

# ==== PREPARE DATA ====
(X_train, y_train), (X_valid, y_valid) = prepare_sequences_targets(train_df, seq_length,
                                      forecast_horizon, cont_vars, target_col, ticker_col, date_col, valid_pct=0.2)

# ==== XGBOOST GRID SEARCH ====
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# Define XGBoost model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

param_dist = {
    'n_estimators': randint(50, 100),           # Narrowed
    'max_depth': randint(3, 5),
    'learning_rate': uniform(0.003, 0.005),     # FastAI suggested ~0.005
    'subsample': uniform(0.8, 0.2),             # 0.8 to 1.0
    'colsample_bytree': uniform(0.8, 0.2),
    'reg_alpha': uniform(0, 0.05),
    'reg_lambda': uniform(1, 2),
}

random_search = RandomizedSearchCV(
    estimator=xgb.XGBRegressor(objective='reg:squarederror', random_state=42),
    param_distributions=param_dist,
    n_iter=10,            # Limit to 10 iterations
    scoring='neg_mean_squared_error',
    cv=2,                 # Use 2-fold CV to cut time
    verbose=1,
    n_jobs=2              # Avoid using all cores
)
# Run tuning
random_search.fit(X_train, y_train)

# Best model
print("Best parameters found: ", random_search.best_params_)
print("Best CV MSE: ", -random_search.best_score_)

best_model = random_search.best_estimator_

# Save model
import joblib
import os

save_dir = '/content/drive/MyDrive/StockModels'
os.makedirs(save_dir, exist_ok=True)
model_path = os.path.join(save_dir, 'xgb_best_model_randomsearch.joblib')
joblib.dump(best_model, model_path)
print(f"Model saved to {model_path}")

# Evaluate
y_valid_pred = best_model.predict(X_valid)
mae_valid = mean_absolute_error(y_valid, y_valid_pred)
mse_valid = mean_squared_error(y_valid, y_valid_pred)
print(f"Validation MAE: {mae_valid:.4f}")
print(f"Validation MSE: {mse_valid:.4f}")
```
